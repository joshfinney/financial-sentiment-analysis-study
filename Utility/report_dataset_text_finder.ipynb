```python
import pandas as pd
from datasets import load_dataset

pd.set_option('display.max_colwidth', None)

# Load the dataset from the hub
dataset = load_dataset("ChanceFocus/fiqa-sentiment-classification")

# Convert splits into dataframes
train_df = pd.DataFrame(dataset['train'])
validation_df = pd.DataFrame(dataset['valid'])
test_df = pd.DataFrame(dataset['test'])

# Concatenate the dataframes
df = pd.concat([train_df, validation_df, test_df], ignore_index=True)

# Display dataframe information
display(df.info())
```

```python
# add column sentiment to the df where sentiment is over 1.3 if the label is positive and under -0.33 if the label is negative else neutral
df['sentiment'] = df['score'].apply(lambda x: 'positive' if x > 0.3333 else ('negative' if x < -0.33333 else 'neutral'))

# rename sentence to text
df.rename(columns={'sentence': 'text'}, inplace=True)
```

```python
# Filter rows based on the length of the text
filtered_df = df[(df['text'].str.len() > 45) & (df['text'].str.len() < 51)]
# display(filtered_df[['text', 'sentiment']])

# Helper function to safely sample and display rows
def safe_sample_display(df, n_samples, sentiment):
    if len(df) < n_samples:
        print(f"Not enough samples for {sentiment} sentiment")
    else:
        display(df.sample(n=n_samples))

# Display samples or a message if there are not enough samples
safe_sample_display(filtered_df[filtered_df['sentiment'] == 'positive'], 3, 'positive')
safe_sample_display(filtered_df[filtered_df['sentiment'] == 'neutral'], 3, 'neutral')
safe_sample_display(filtered_df[filtered_df['sentiment'] == 'negative'], 3, 'negative')
```
