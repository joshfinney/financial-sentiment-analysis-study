```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

def plot_conf_mats(cms, model):
    class_names = ['Negative', 'Neutral', 'Positive']  # Names of classes
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4.5))
    
    for ax, (feature_representation, cm) in zip(axes, cms.items()):
        sns.heatmap(cm, annot=True, cmap="YlGnBu", fmt='g', ax=ax)
        ax.set_title(f'Confusion Matrix\n{model} using {feature_representation}', fontsize=12, pad=10)
        ax.set_xlabel('Predicted label', labelpad=5)
        ax.set_ylabel('Actual label', labelpad=5)
        tick_marks = np.arange(len(class_names)) + 0.5
        ax.set_xticks(tick_marks)
        ax.set_xticklabels(class_names)
        ax.set_yticks(tick_marks)
        ax.set_yticklabels(class_names)
    
    plt.subplots_adjust(wspace=6)  # adjust the width space between subplots

    plt.tight_layout()
    plt.savefig(f"Figs/{model}_DS1_Comparison_Confusion_Matrix.pdf")
    plt.show()

class_names = ['Negative', 'Neutral', 'Positive']

# Sample data for confusion matrices
cms_data = {
    "Logistic Regression": {
        "TF-IDF": pd.DataFrame([[332, 75, 18], [79, 2552, 56], [17, 116, 1165]], columns=class_names, index=class_names),
        "BoW": pd.DataFrame([[309, 100, 16], [29, 2616, 42], [7, 97, 1194]], columns=class_names, index=class_names)
    },
    "Decision Tree": {
        "TF-IDF": pd.DataFrame([[315, 48, 26], [66, 2427, 91], [11, 73, 1216]], columns=class_names, index=class_names),
        "BoW": pd.DataFrame([[315, 56, 18], [37, 2470, 77], [14, 57, 1229]], columns=class_names, index=class_names)
    },
    "Random Forest": {
        "TF-IDF": pd.DataFrame([[306, 69, 14], [21, 2516, 47], [5, 91, 1204]], columns=class_names, index=class_names),
        "BoW": pd.DataFrame([[315, 60, 14], [19, 2514, 51], [7, 77, 1216]], columns=class_names, index=class_names)
    },
}

# Plot confusion matrices for each model
for model, cms in cms_data.items():
    plot_conf_mats(cms, model)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

def plot_conf_mat(cm, model):
    class_names = ['Negative', 'Neutral', 'Positive']  # Names of classes
    fig, ax = plt.subplots(figsize=(5, 4.5))  # Single subplot
    
    sns.heatmap(cm, annot=True, cmap="YlGnBu", fmt='g', ax=ax)
    ax.set_title(f'Confusion Matrix\n{model}', fontsize=12, pad=10)
    ax.set_xlabel('Predicted label', labelpad=5)
    ax.set_ylabel('Actual label', labelpad=5)
    tick_marks = np.arange(len(class_names)) + 0.5
    ax.set_xticks(tick_marks)
    ax.set_xticklabels(class_names)
    ax.set_yticks(tick_marks)
    ax.set_yticklabels(class_names)
    
    plt.tight_layout()
    plt.savefig(f"Figs/{model}_DS1_Confusion_Matrix.pdf")
    plt.show()

class_names = ['Negative', 'Neutral', 'Positive']

# Sample data for confusion matrices
cms_data = {
    "Convolutional Neural Network (CNN)": pd.DataFrame([[348, 34, 7], [17, 2537, 30], [3, 49, 1248]], columns=class_names, index=class_names),
}

# Plot confusion matrices for each model
for model, cm in cms_data.items():
    plot_conf_mat(cm, model)  # The feature representation is just an example label
```
